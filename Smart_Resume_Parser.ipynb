```python
from transformers import pipeline
```

```python
# Load NER pipeline
ner = pipeline(
    "ner",
    model="dbmdz/bert-large-cased-finetuned-conll03-english",
    aggregation_strategy="simple"
)
```

```python
resume_text = """
Satwik is a final year student at PES University.
He completed an internship at JP Morgan and later worked with Google as a Software Engineer.
"""
```

```python
entities = ner(resume_text)
```

```python
print("Extracted Entities:")
for ent in entities:
    print(f"{ent['word']}  -->  {ent['entity_group']}  (confidence: {ent['score']:.2f})")
```
