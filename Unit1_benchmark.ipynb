```python
from transformers import pipeline
```

```python
bert_gen = pipeline("text-generation", model="bert-base-uncased")
roberta_gen = pipeline("text-generation", model="roberta-base")
bart_gen = pipeline("text-generation", model="facebook/bart-base")
```

**Text Generation**

```python
prompt = "The future of Artificial Intelligence is"

print("BERT OUTPUT:")
print(bert_gen(prompt, max_length=30))

print("\nRoBERTa OUTPUT:")
print(roberta_gen(prompt, max_length=30))

print("\nBART OUTPUT:")
print(bart_gen(prompt, max_length=30))
```

**FILL-MASK (MLM)**

```python
bert_mask = pipeline("fill-mask", model="bert-base-uncased")
roberta_mask = pipeline("fill-mask", model="roberta-base")
bart_mask = pipeline("fill-mask", model="facebook/bart-base")
```

```python
bert_sentence = "The goal of Generative AI is to [MASK] new content."

print("BERT MASK:")
print(bert_mask(bert_sentence))
```

```python
roberta_sentence = "The goal of Generative AI is to <mask> new content."

print("RoBERTa MASK:")
print(roberta_mask(roberta_sentence))
```

```python
bart_sentence = "The goal of Generative AI is to <mask> new content."

print("BART MASK:")
print(bart_mask(bart_sentence))
```

**QUESTION ANSWERING**

```python
bert_qa = pipeline("question-answering", model="bert-base-uncased")
roberta_qa = pipeline("question-answering", model="roberta-base")
bart_qa = pipeline("question-answering", model="facebook/bart-base")
```

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

print("BERT QA:")
print(bert_qa(question=question, context=context))

print("\nRoBERTa QA:")
print(roberta_qa(question=question, context=context))

print("\nBART QA:")
print(bart_qa(question=question, context=context))
```
